```r
#Importing libraries

library(tidyverse)
library(dplyr)
library(lubridate)
library(tidyr)
library(Hmisc)
library(gapminder)
```

```r
#loading data tables
website_pageviews <- read.csv("C:/Users/DELL/Downloads/website_pageviews.csv")
website_sessions <- read.csv("C:/Users/DELL/Downloads/website_sessions.csv")
products <- read.csv("C:/Users/DELL/Downloads/products.csv")
orders <- read.csv("C:/Users/DELL/Downloads/orders.csv")
order_item <- read.csv("C:/Users/DELL/Downloads/order_item.csv")
order_item_refund <- read.csv("C:/Users/DELL/Downloads/order_item_refund.csv")

```

```r
#Data Wrangling and Renaming

website_pageviews <- website_pageviews |> rename(pageview_timestamp = created_at)
website_sessions <- website_sessions |> rename(websession_timestamp = created_at)
webdetails <- full_join(website_pageviews,website_sessions,by = "website_session_id")
webdetails <- webdetails |> rename(webuser_id = user_id)
 orders <- orders |> rename(order_timestamp = created_at)
 order_item <- order_item |> rename(orderitem_timestamp = created_at)
 order_item <- order_item |> rename(orderitem_price = price_usd)
 orders <- orders |> rename(order_price = price_usd)
 orders <- orders |> rename(order_cogs_usd = cogs_usd)
 order_item <- order_item |> rename(orderitem_cogs_usd = cogs_usd)
join2 <- full_join(orders,order_item)
 order_item_refund <- order_item_refund |> rename(refund_timestamp = created_at)
join1 <- full_join(order_item_refund,join2)
 products <- products |> rename(product_list_timestamp = created_at)
 ORD <- full_join(join1,products)
ORD <- ORD |> rename(orduser_id = user_id)
maven_fuzzy <- full_join(ORD,webdetails,by = "website_session_id") 


```

# Refunds Analysis

```r
#refund percentage for each year and quarter
 maven_fuzzy |>   
  group_by(year(orderitem_timestamp),quarter(orderitem_timestamp)) |> 
  summarise(Refund_percent = round(n_distinct(order_item_refund_id,na.rm = TRUE)/n_distinct(order_item_id,na.rm = TRUE)*100)) |> 
  view()
#return trigger in each quarter
JJ <- maven_fuzzy |> 
  filter(is.na(order_item_refund_id)) |> 
  group_by(quarter = quarter(order_timestamp)) |> 
  summarise(sold = n_distinct(order_id)) |> left_join(
    maven_fuzzy |> 
      filter(!is.na(order_item_refund_id)) |> 
      group_by(quarter = quarter(order_timestamp)) |> 
      summarise(return = n_distinct(order_id))  
  ) |> group_by(quarter) |> 
  summarise(return_trigger = round(sold/return)) |> view()     
#average days taken to generate return
KK <- maven_fuzzy |> 
  filter(!is.na(order_item_refund_id)) |> 
  group_by(order_id) |> 
  mutate(TFR = difftime(refund_timestamp,order_timestamp)) |>
  ungroup() |> 
  summarise(median(TFR)) |> view()
```

# Products Analysis


```r
#Units sold per product in last year(2015)
LL <- ORD |>   
  mutate(year = year(orderitem_timestamp),quarter = quarter(orderitem_timestamp)) |>
  filter(year == "2015") |> 
  group_by(product_name,year) |> 
  summarise(units_sold = n_distinct(order_item_id)) |> view()

#Units sold per product in each quarter
MM <- ORD |>   
  mutate(year = year(orderitem_timestamp),quarter = quarter(orderitem_timestamp)) |> 
  group_by(product_name,year,quarter) |> 
  summarise(unit_sold = n_distinct(order_item_id)) |>
  group_by(product_name,quarter) |> 
  summarise(unit_solds = round(mean(unit_sold))) |> 
  view()

#orders_percenatge_as_primary_product
OO <- ORD |>
  group_by(ProductID = primary_product_id) |> 
  summarise(A = n_distinct(order_id)) |> left_join(
    ORD |>
      group_by(ProductID = product_id) |> 
      summarise(B = n_distinct(order_id))) |> 
  mutate(orders_percenatge_as_primary_product = (A/B)*100) |> view()  

#products orders with multiple and single items
PP <- ORD |> filter(items_purchased == "2") |> 
  group_by(product_name) |> 
  summarise(MPO = n_distinct(order_id)) |> left_join(
    ORD |> 
      group_by(product_name) |>
      summarise(SPO = n_distinct(order_id))) |>
  group_by(product_name) |> 
  summarise(MPO_percentage = (MPO/SPO)*100) |> view()

#Basketing of products
QQ <- ORD |> 
  group_by(primary_product_id,product_id) |> 
  summarise(Orders = n_distinct(order_id)) |> 
  filter(!(primary_product_id == product_id)) |> 
  group_by(primary_product_id) |> 
  slice_max(Orders) |> view()

#Frequent shopping hour for each product
RR <- maven_fuzzy |> 
  mutate(OH = hour(order_timestamp)) |> 
  group_by(product_name) |> 
  summarise(frequent_shopping_hour = which.max(table(OH))) |> view()

#Customer retention percentage by product
R1 <- maven_fuzzy |> 
  group_by(webuser_id = orduser_id,na.rm = TRUE) |> 
  summarise(order_id = min(order_id,na.rm = TRUE)) |> 
  left_join(order_item) |> 
  filter(!is.na(order_id)) |> 
  filter(is_primary_item == "1") |> 
  select(webuser_id,
         FP = product_id,
         FPTS = orderitem_timestamp)
R2 <- full_join(maven_fuzzy,R1) 

R3 <- R2 |> 
  filter(websession_timestamp>FPTS || order_timestamp==FPTS) |> 
  filter(is.na(order_item_refund_id)) |> 
  group_by(FP,webuser_id) |> 
  summarise(SES = n_distinct(website_session_id)) |> 
  filter(SES>1) |> 
  group_by(FP) |> 
  summarise(Repeat_session = n_distinct(webuser_id)) |> full_join(
    R2 |> 
      filter(websession_timestamp>FPTS || order_timestamp==FPTS) |> 
      group_by(FP,webuser_id) |> 
      summarise(SES = n_distinct(website_session_id)) |> 
      filter(SES=="1") |> 
      group_by(FP) |> 
      summarise(No_repeat_session = n_distinct(webuser_id))) |> 
  group_by(FP) |> 
  summarise(Customer_retention_by_product = (Repeat_session/(Repeat_session+No_repeat_session))*100)

```
